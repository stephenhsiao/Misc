```python
# General imports
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import csv
import seaborn as sns
```

```python
def graph(x_range):  
    # set the x ranges
    x = np.array(x_range) 
    
    # set the function
    # y = np.power(1.0+1.0/x,x)
    # Logistic function
    y = 1 / (1 + np.exp(-x))
    
    # configurations
    fig = plt.figure(figsize=(10, 8), dpi=100)
    plt.plot(x, y)
    plt.title('LineChart', fontsize=16)
    plt.xlabel('n')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()
```

```python
# plot the function
graph(range(-20, 20))
```
