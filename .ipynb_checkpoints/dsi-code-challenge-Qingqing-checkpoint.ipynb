```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# make sure plots will be shown inline
%matplotlib inline
```

## 1. Read in the data

```python
url = 'https://s3-us-west-2.amazonaws.com/ga-dat-2015-suneel/datasets/breast-cancer.csv'
```

```python
#column 0 is the ID of a patient record
cancer = pd.read_csv(url,  index_col=0)

#set column names
names = ['radius', 'texture', 'perimeter', 'area', 'smoothness',
                  'compactness', 'concavity', 'concave points', 'symmetry', 'fractal dimension']
names_mean = [x+"_mean" for x in names]
names_SD =  [x+"_SD" for x in names]
names_worst =  [x+"_worst" for x in names]
cancer.columns =  ['Diagnosis'] + names_mean + names_SD + names_worst
```

```python
cancer.shape
```

```python
# Information about the data
# cancer.shape
# cancer.describe
cancer
```

## 2. Separate the data into feature and target.

```python
myList = [i for i in range(len(cancer.columns)) if i != 0]
X = cancer.iloc[:, myList]
y = cancer.iloc[:,0]

print y.describe()
```

```python
X.head(5)
```

```python
y.head(5)
```

## 3. Create and evaluate using cross_val_score and 5 folds.
- What is the mean accuracy?
 - Answer: the rate of correctly predicted instances is called accuracy. Mean accuracy is the average of all accuracy
 scores coming from k cross validation models.
- What is the standard deviation of accuracy?
 - Answer: standard deviation is used to measure how spread out the accuracy scores are. It's defined as the square root of the
 variance, which is defined as the average of the squared differences from the mean. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import cross_val_score
from sklearn import metrics
```

```python
# This dataset has all numeric features, logistic regression should be a good learning algorithm to start with. 
logit = LogisticRegression()
scores = cross_val_score(logit, X, y, cv=5, scoring='accuracy')
```

```python
print scores
print scores.mean()
print scores.std()
```

## 4. Get a classification report to identify type 1, type 2 errors.
- Use train_test_split to run your model once, with a test size of 0.33
- Make predictions on the test set
- Compare the predictions to the answers to determine the classification report

```python
# use train/test split to run the model
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=77)

# make prediction on the test set
logit.fit(X_train, y_train)
y_pred = logit.predict(X_test)

# check classification accuracy
print metrics.accuracy_score(y_test, y_pred)
```

```python
# print confusion matrix 
print metrics.confusion_matrix(y_test, y_pred)
```

```python
# print classification report
print metrics.classification_report(y_test, y_pred)
```

## 5. Scale the data and see if that improves the score.

```python
# standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# run cross_val_score using scaled data
scores = cross_val_score(logit, X_scaled, y, cv=5, scoring='accuracy')
print scores
print scores.mean()
print scores.std()

'''there might be a slight problem with this, since X got scaled before the cross validation starts to split the data into k
folds. A better solution is to use pipeline to work around the problem, which might not be the focus of this exercise.
I will skip for now.'''
```

## 6. Tune the model using automated parametric grid search via LogisticRegressionCV. Explain your intution behind what is being tuned.

### Q: What should we do to prevent overfitting so our model generalizes well to the test data?

```python
from sklearn.linear_model import LogisticRegressionCV

# C is a regularization factor here to prevent overfitting. C is actually inverse of regularization parameter. 
# smaller values specify stronger regularization.  
searchCV = LogisticRegressionCV(
        Cs=list(np.power(10.0, np.arange(-10, 10)))
        ,scoring='accuracy'
        ,cv=5
    )

searchCV.fit(X, y)
```

```python
# Compute the mean accuracy. We get 0.95606, which is slightly better than what we got from step 3 above (the mean accuracy score
# was 0.95079).
np.array([row[10] for row in searchCV.scores_['M']]).mean()
```

### Q: What was the best C?

```python
searchCV.C_
```
